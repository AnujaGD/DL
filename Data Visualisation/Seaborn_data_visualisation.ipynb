```python
#import required libraries and dataset

import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
countries_path = "D:/factbook.csv"
```

```python
#get the dataset as pandas dataframe
countries_data = pd.read_csv(countries_path,index_col="Country")
```

```python
#deleted 1st empty row
countries_data = countries_data.iloc[1:]
```

# Plot bar graph of Country vs Fertility rate using seaborn 

```python
plt.figure(figsize=(50,14))
sb.barplot(data=countries_data,y='Total fertility rate(children born/woman)',x=countries_data.index)
plt.xticks(rotation=90)
```

# Plot histogram to see how population varies

```python
plt.figure(figsize=(14,6))
sb.distplot(a =countries_data[ "Population"].dropna(),kde=False)
```

# Plot 2D histogram(joint plot) to see variations in Birth Rate and Death Rate

```python
plt.figure(figsize=(14,6))
sb.jointplot(data=countries_data,y='Birth rate(births/1000 population)',x="Death rate(deaths/1000 population)",kind="kde")
```

```python
plt.figure(figsize=(14,6))
sb.scatterplot(data=countries_data,y='Birth rate(births/1000 population)',x="Death rate(deaths/1000 population)")
```

# regplot is used to find a linear relation between variables

```python
plt.figure(figsize=(14,6))
sb.regplot(data=countries_data,y='Birth rate(births/1000 population)',x="Death rate(deaths/1000 population)")
```

```python
plt.figure(figsize=(14,6))
sb.scatterplot(data=countries_data,y='Electricity - consumption(kWh)',x="Electricity - production(kWh)")
```

```python
plt.figure(figsize=(14,6))
sb.lmplot(data=countries_data,y='Electricity - consumption(kWh)',x="Electricity - production(kWh)")
```

# Scatter plot with hue is used to see how three variables are correlated 


```python
plt.figure(figsize=(14,6))
sb.scatterplot(data=countries_data,y='Birth rate(births/1000 population)',x="Death rate(deaths/1000 population)",hue="Unemployment rate(%)")
```

```python
plt.figure(figsize=(14,6))
sb.scatterplot(data=countries_data,y='Exports',x="GDP")
```

```python
plt.figure(figsize=(14,6))
sb.regplot(data=countries_data,y='Exports',x="GDP")
```

# kde plot (kernel density plot) is a smooth form of Histogram.

```python
plt.figure(figsize=(14,6))
sb.kdeplot(data=countries_data['GDP'],shade=True)
```
